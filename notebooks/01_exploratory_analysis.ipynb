# Forecasting Case Study - Exploratory Analysis

This notebook documents the basic EDA workflow for the sales forecasting assignment.

```python
import pandas as pd

DATA_PATH = '../data/Forecasting Case- Study.xlsx'
df = pd.read_excel(DATA_PATH)
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
summary = {
    'rows': len(df),
    'states': df['State'].nunique(),
    'start_date': df['Date'].min(),
    'end_date': df['Date'].max(),
    'categories': df['Category'].unique().tolist(),
}
summary
```

```python
weekly = (
    df.set_index('Date')
      .groupby('State')['Total']
      .resample('W-SUN')
      .sum(min_count=1)
      .reset_index()
)
weekly.head()
```
